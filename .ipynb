```python
def add(x, y):
    return x + y

def subtraction(x, y):
    return x - y

def multiplication(x, y):
    return x * y

def divide(x, y):
    return x / y

while True:
    print("Please select operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Quit")

    choice = input("Enter choice (1/2/3/4/5): ")

    if choice in ('1', '2', '3', '4'):
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))

        if choice == '1':
            print(num1, "+", num2, "=", add(num1, num2))

        elif choice == '2':
            print(num1, "-", num2, "=", subtraction(num1, num2))

        elif choice == '3':
            print(num1, "*", num2, "=", multiplication(num1, num2))

        elif choice == '4':
            print(num1, "/", num2, "=", divide(num1, num2))
    elif choice == '5':
        print("Goodbye!")
        break
    else:
        print("Invalid input")
```

```python
def add(x, y):
    return x + y

def subtraction(x, y):
    return x - y

def multiplication(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "zero is not divided"
    else:
       return x / y

while True:
    choice = input("enter choice (q to quit): ")

    if choice == "q":
        print("Goodbye")
        break
    
    if choice in ('+', '-', '*', '/'):
        num_count = int(input("how many numbers do you want to calculate? "))
        nums = []
        for i in range(num_count):
            num = float(input(f"enter number {i+1}: "))
            nums.append(num)
        
        if choice == "+":
            result = nums[0]
            for i in range(1, len(nums)):
                result = add(result, nums[i])
            print("result =", result)
        elif choice == "-":
            result = nums[0]
            for i in range(1, len(nums)):
                result = subtraction(result, nums[i])
            print("result =", result)
        elif choice == "*":
            result = nums[0]
            for i in range(1, len(nums)):
                result = multiplication(result, nums[i])
            print("result =", result)
        elif choice == '/':
            result = nums[0]
            for i in range(1, len(nums)):
                result = divide(result, nums[i])
            print("result =", result)
    else:
        print("Invalid operation selection")
```
